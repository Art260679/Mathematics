## __1.__ Решить систему уравнений методом Гаусса:

$$\begin{cases}
x_{1}+x_{2}-x_{3}-2x_{4}=0, \\
2x_{1}+x_{2}-x_{3}+x_{4}=-2, \\
x_{1}+x_{2}-3x_{3}+x_{4}=4.
\end{cases}$$

##### Решение

Приведем расширенную матрицу системы к ступенчатому виду:

$\begin{pmatrix}
1 & 1 & -1 & -2 & 0 \\
2 & 1 & -1 & 1 & -2 \\
1 & 1 & -3 & 1 & 4
\end{pmatrix}=\begin{pmatrix}
1 & 1 & -1 & -2 & 0 \\
0 & -1 & 1 & 5 & -2 \\
1 & 1 & -3 & 1 & 4
\end{pmatrix}=
\begin{pmatrix}
1 & 1 & -1 & -2 & 0 \\
0 & -1 & 1 & 5 & -2 \\
0 & 0 & -2 & 3 & 4
\end{pmatrix}$

$\begin{cases}
x_{1}+x_{2}-x_{3}-2x_{4}=4, \\
-x_{2}+x_{3}+5x_{4}=-2, \\
-2x_{3}+3x_{4}=4;
\end{cases}$

Из уравнения 3 системы  найдем переменную $x_{3}$: <br>
$x_3=-2+\frac{3}{2}x_4$

Из уравнения 2 системы найдем переменную $x_2$: <br>
$x_2=\frac{13}{2}x_4$

Из уравнения 1 системы найдем переменную $x_1$: <br>
$x_1=-x_2+x_3+2x_4=-(\frac{13}{2}x_4)+(-2+\frac{3}{2}x_4)+2x_4=-2-3x_4$

Ответ: <br>
$ x_1=-2-3x_4 $<br>
$x_2=\frac{13}{2}x_4$<br>
$x_3=-2+\frac{3}{2}x_4$<br>
$x_4=x_4$

###### Проверка:

```python
import numpy as np
```

```python
def check(A, B):
    a_r, b_r, n = np.linalg.matrix_rank(A), np.linalg.matrix_rank(B), A.shape[1]
    if  a_r == b_r:
        if a_r == n:
            print(f'Проверка совместности системы: \nrank(A) = {a_r}, rank(B) = {b_r}, ' 
                  f'т.к. ранг матрицы = рангу расширенной матрицы = количество переменных({n}), '
                  f'\nто система совместна и решение одно.')
        elif a_r < n:
            print(f'Проверка совместности системы: \nrank(A) = {a_r}, rank(B) = {b_r}, количество переменных = {n}, ' 
                  f'т.к. ранг матрицы = ранг расширенной матрицы < количества переменных,'
                  f'\nто система неопределена, т.е. количество решений бесконечно.')
    elif a_r < b_r:
        print(f'Проверка совместности системы: \nrank(A) = {a_r}, rank(B) = {b_r}, ' 
              f'т.к. ранг матрицы < ранг расширенной матрицы, то система несовместна.')
```

```python
A = np.array([[3, -1, 1], [2, -5, -3], [1, 1, -1]])
B = np.array([[3, -1, 1, 4], [2, -5, -3, -17], [1, 1, -1, 0]])
check(A, B)
```

```python
A = np.array([[1, 2, 5], [3, 1, -8]])
B = np.array([[1, 2, 5, 4], [3, 1, -8, -2]])
check(A, B)
```
